```python
class Node:
    def __init__(self,data):
        self.data=data
        self.next=None
class Linked:
    def __init__(self):
        self.start=None
    def insert(self,data):
        new_node=Node(data)
        if self.start==None:
            self.start=new_node
            
        else:
            temp=self.start
            while temp.next!=None:
                temp=temp.next
            temp.next=new_node
            
    def pti(self):
        if self.start==None:
            print('nothing in list')
        else:
            temp=self.start
            while temp!=None:
                print(temp.data,end=' ')
                temp=temp.next
    def reverse(self):
        cur=self.start
        prev=None
        while cur:
            nxt=cur.next
            cur.next=prev
            prev=cur
            cur=nxt
        self.start=prev
                
L=Linked()
L.insert('A')
L.insert('B')
L.insert('C')
L.insert('D')
L.reverse()
L.pti()
```
